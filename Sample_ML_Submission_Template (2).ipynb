## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ================================
# DATA WRANGLING / PREPROCESSING
# ================================

# Make a copy of dataset
# ================================
# DATA WRANGLING / PREPROCESSING
# ================================

# Make a copy of dataset
df_processed = df.copy()

# --------------------------------
# 1. STANDARDIZE COLUMN NAMES
# --------------------------------
df_processed.columns = (
    df_processed.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.lower()
)

# --------------------------------
# 2. REMOVE DUPLICATES
# --------------------------------
df_processed = df_processed.drop_duplicates()

# --------------------------------
# 3. CONVERT DATE COLUMNS TO DATETIME
# --------------------------------

date_cols = ['order_date_time', 'issue_reported_at',
             'issue_responded', 'survey_response_date']

for col in date_cols:
    if col in df_processed.columns:
        df_processed[col] = pd.to_datetime(df_processed[col], errors='coerce')

# --------------------------------
# 4. HANDLE MISSING VALUES
# --------------------------------

# Numerical columns
num_cols = df_processed.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df_processed[col] = df_processed[col].fillna(df_processed[col].median())

# Categorical columns
cat_cols = df_processed.select_dtypes(include=['object']).columns

for col in cat_cols:
    df_processed[col] = df_processed[col].fillna(df_processed[col].mode()[0])

# --------------------------------
# 5. FEATURE ENGINEERING
# --------------------------------

if all(col in df_processed.columns for col in date_cols):

    df_processed['response_time'] = (
        (df_processed['issue_responded'] - df_processed['issue_reported_at'])
        .dt.total_seconds() / 60
    )

    df_processed['resolution_time'] = (
        (df_processed['survey_response_date'] - df_processed['issue_reported_at'])
        .dt.total_seconds() / 60
    )

    df_processed['order_processing_time'] = (
        (df_processed['issue_reported_at'] - df_processed['order_date_time'])
        .dt.total_seconds() / 60
    )

    # Remove negative time values (if any)
    time_cols = ['response_time','resolution_time','order_processing_time']
    for col in time_cols:
        df_processed[col] = df_processed[col].apply(lambda x: x if x >= 0 else 0)

# --------------------------------
# 6. DROP IRRELEVANT COLUMNS
# --------------------------------

drop_cols = ['unique_id', 'order_id', 'agent_name',
             'supervisor', 'manager']

df_processed = df_processed.drop(
    columns=[col for col in drop_cols if col in df_processed.columns]
)

# --------------------------------
# 7. HANDLE OUTLIERS USING IQR METHOD
# --------------------------------

for col in ['connected_handling_time', 'item_price']:
    if col in df_processed.columns:

        Q1 = df_processed[col].quantile(0.25)
        Q3 = df_processed[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df_processed[col] = np.where(
            df_processed[col] < lower, lower, df_processed[col]
        )
        df_processed[col] = np.where(
            df_processed[col] > upper, upper, df_processed[col]
        )

# Rename for consistency
df = df_processed.copy()

# --------------------------------
# 8. FINAL CHECK
# --------------------------------

print("Final Dataset Shape:", df.shape)
print("\nMissing Values After Cleaning:\n", df.isnull().sum())
df.head()
````

Q3 (75th percentile)

Lower Bound = Q1 - 1.5 × IQR

Values beyond this range are considered outliers.

I applied the IQR approach for the following reasons:

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Handling Outliers using IQR Method
def treat_outliers_iqr(df, column):

    if column in df.columns:

        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Capping (Winsorization)
        df[column] = df[column].clip(lower_bound, upper_bound)

    return df


# Apply to numerical columns
numerical_cols = ['response_time', 'connected_handling_time']

for col in numerical_cols:
    df = treat_outliers_iqr(df, col)

print("Outliers treated successfully.")
````

I handled outliers using capping (winsorization) rather than removing them completely to avoid losing data.
Dealing with outliers is essential for improving model robustness and avoiding the distortion of statistical summaries.

#### 1. Expand Contraction

````python
# Expand Contraction
# Expand Contractions

import re

def clean_text(text):

    if pd.isnull(text):
        return ""

    text = str(text).lower()

    # Expand contractions
    contractions = {
        "can't": "cannot",
        "won't": "will not",
        "n't": " not",
        "'re": " are",
        "'s": " is",
        "'d": " would",
        "'ll": " will",
        "'ve": " have",
        "'m": " am"
    }

    for key, value in contractions.items():
        text = text.replace(key, value)

    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Apply on correct column name
if 'customer_remarks' in df.columns:
    df['customer_remarks'] = df['customer_remarks'].apply(clean_text)

print("Text cleaning completed successfully.")
````

#### 6. Rephrase Text

````python
# Rephrase Text
import re

def rephrase_text(text):

    if pd.isnull(text):
        return ""

    text = str(text).lower()

    # Expand common contractions
    contractions = {
        "can't": "cannot",
        "won't": "will not",
        "n't": " not",
        "'re": " are",
        "'s": " is",
        "'d": " would",
        "'ll": " will",
        "'ve": " have",
        "'m": " am"
    }

    for key, value in contractions.items():
        text = text.replace(key, value)

    # Replace informal words
    replacements = {
        "pls": "please",
        "plz": "please",
        "asap": "as soon as possible",
        "btw": "by the way",
        "u": "you",
        "ur": "your",
        "thx": "thank you"
    }

    words = text.split()
    words = [replacements.get(word, word) for word in words]

    text = " ".join(words)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Apply to correct column
if 'customer_remarks' in df.columns:
    df['customer_remarks'] = df['customer_remarks'].apply(rephrase_text)

print("Text rephrasing completed successfully.")
````

### 7. Dimesionality Reduction

##### Do you think that dimensionality reduction is needed? Explain Why?

Answer Here.Given that we've already performed feature selection and reduced our dataset to 20 features, further dimensionality reduction is likely not strictly needed at this stage.

Here's why:

Manageable Number of Features: Twenty features is generally considered a low-to-moderate number for machine learning models. Most algorithms can handle this many features efficiently without significant computational burden or high risk of overfitting, especially with proper regularization and cross-validation.
Potential Loss of Information: Dimensionality reduction techniques (like PCA) work by transforming the original features into a smaller set of new, uncorrelated features. While this can be beneficial for very high-dimensional data, with only 20 features, applying further reduction might lead to a loss of potentially valuable information contained within those features, which could negatively impact model performance.
Interpretability: The features we selected are often directly interpretable (e.g., response_time, tenure_bucket, specific category or sub-category). Reducing their dimensionality further would transform them into abstract components, making it harder to interpret why a model makes a certain prediction. For business applications, interpretability is often crucial.
While dimensionality reduction can be very powerful for datasets with hundreds or thousands of features, for a dataset already refined to 20 features, the benefits often do not outweigh the potential drawbacks of information loss and reduced interpretability.

##### Which dimensionality reduction technique have you used and why? (If dimensionality reduction done on dataset.)

Answer Here.I applied PCA (Principal Component Analysis) for dimensionality reduction. Here’s why:

Variance Preservation: PCA is used to transform the original features into a new set of orthogonal (uncorrelated) components, ranked according to the amount of variance they explain. By specifying n_components=0.95, I sought to retain 95% of the total variance in the dataset, thereby ensuring that most of the key information is retained despite the dimensionality reduction.
Mitigation of the Curse of Dimensionality: Even though our feature space was reduced to 20 features using SelectKBest, PCA can still be applied, particularly if there is multicollinearity among the remaining features. PCA assists in producing a more concise representation of the dataset, which can result in faster training times and possibly improved generalization for certain models.
Noise Reduction: By emphasizing the components that explain the most variance, PCA can implicitly remove some noise in the less important components, thereby possibly improving the robustness of models.

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
# ==================================
# Hyperparameter Tuning using GridSearchCV
# ==================================

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score

# Define the parameter grid for Logistic Regression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization parameter
    'solver': ['liblinear', 'lbfgs'], # Algorithm to use in the optimization problem
    'penalty': ['l1', 'l2'] # Specify the norm of the penalty
}

# Initialize GridSearchCV
# Using 'f1' as scoring since it's good for imbalanced datasets
grid_search = GridSearchCV(
    LogisticRegression(max_iter=1000, random_state=42),
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1, # Use all available CPU cores
    verbose=1
)

# Fit GridSearchCV to the SMOTE-resampled training data
grid_search.fit(X_train_smote, y_train_smote)

# Get the best estimator
best_log_model = grid_search.best_estimator_

# Predict using the best model on the test set
y_pred_best = best_log_model.predict(X_test)

# Evaluate the best model
accuracy_best_log = accuracy_score(y_test, y_pred_best)
precision_best_log = precision_score(y_test, y_pred_best)
recall_best_log = recall_score(y_test, y_pred_best)
f1_best_log = f1_score(y_test, y_pred_best)

print("Best Parameters:", grid_search.best_params_)
print("Accuracy with Best Model:", accuracy_best_log)
print("F1 Score with Best Model:", f1_best_log)
print("\nClassification Report with Best Model:\n", classification_report(y_test, y_pred_best))
print("\nConfusion Matrix with Best Model:\n", confusion_matrix(y_test, y_pred_best))
````
