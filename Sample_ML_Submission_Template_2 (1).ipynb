## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Data Wrangling Code
df = df.drop_duplicates()

date_columns = ['order_date_time',
                'Issue_reported at',
                'issue_responded',
                'Survey_response_Date']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')




df['response_time_minutes'] = (
    (df['issue_responded'] - df['Issue_reported at'])
    .dt.total_seconds() / 60
)


df['order_month'] = df['order_date_time'].dt.month
df['order_day'] = df['order_date_time'].dt.day
df['order_hour'] = df['order_date_time'].dt.hour


for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)


columns_to_drop = ['Unique id', 'Order_id',
                   'Agent_name', 'Supervisor', 'Manager']

df = df.drop(columns=columns_to_drop, errors='ignore')


from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col])



X = df.drop('CSAT Score', axis=1)
y = df['CSAT Score']

print("Data Wrangling Completed Successfully!")
print("Final Dataset Shape:", df.shape)
````

### What all manipulations have you done and insights you found?

Data Manipulations Performed
Duplicate Removal: Removed duplicate data to ensure consistency in the data and to prevent any biases during model training.
Date-Time Conversion: Converted all the columns that were in date format (order_date_time, Issue_reported at, issue_responded, Survey_response_Date) to the correct datetime format to extract features based on time.
Feature Engineering: Created a new feature named response_time_minutes using the time difference between issue reporting and issue response.
Extracted features such as order month, order day, and order hour to identify patterns based on time.
Missing Value Treatment
Numerical Variables: Imputed numerical values using the median to reduce the effect of extreme values.
Categorical Variables: Imputed categorical values using the mode.
Dropping Irrelevant Features:
Dropped high-cardinality and identifier variables like Unique id, Order_id, Agent_name, Supervisor, and Manager as they do not contribute to the model.
Categorical Encoding:
Used Label Encoding to convert categorical variables into a numerical format.
Feature-Target Separation:
Defined the target variable as CSAT Score and separated the independent variables (X) and dependent variables (y).

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
import numpy as np
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("Outlier treatment completed.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were addressed using the Interquartile Range (IQR) approach. This approach identifies extreme values that are beyond the set range of first quartile (Q1) and third quartile (Q3).

Any value below Q1 – 1.5 * IQR or above Q3 + 1.5 * IQR is considered an outlier. These extreme values were either capped or removed to prevent them from affecting the analysis and machine learning model.

Addressing these outliers improves model stability and minimizes bias, ensuring that model predictions are not dominated by these abnormal values.

````python
# Vectorizing Text
from sklearn.feature_extraction.text import TfidfVectorizer

# Join the lemmatized tokens back into strings for TF-IDF
df['processed_text'] = df['lemmatized_text'].apply(lambda x: ' '.join(x))

# Filter out documents that are empty after all the preprocessing steps
# The error "empty vocabulary" occurs because TfidfVectorizer cannot build a vocabulary
# if all input documents are empty or contain only non-informative tokens that get filtered out.
# This often happens when text preprocessing is too aggressive or applied to data that
# was already not pure text (e.g., label encoded numerical values treated as strings).
meaningful_processed_texts = df[df['processed_text'].str.strip().astype(bool)]['processed_text']

if not meaningful_processed_texts.empty:
    # Initialize TF-IDF Vectorizer
    tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limiting features to avoid high dimensionality

    # Fit and transform only the meaningful processed texts
    tfidf_features = tfidf_vectorizer.fit_transform(meaningful_processed_texts)

    # Convert to DataFrame (optional, but good for inspection)
    # Align the new tfidf_df with the original DataFrame's index
    tfidf_df_temp = pd.DataFrame(tfidf_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=meaningful_processed_texts.index)

    # Create a full tfidf_df DataFrame, filling missing rows (those with empty texts) with zeros
    tfidf_df = pd.DataFrame(0.0, index=df.index, columns=tfidf_vectorizer.get_feature_names_out())
    tfidf_df.update(tfidf_df_temp)

    print("Text Vectorization using TF-IDF Completed. Filtered out empty documents.")
else:
    print("Warning: All processed texts are empty or contain no meaningful vocabulary. TF-IDF vectorization skipped.")
    # Create an empty DataFrame with the original index to maintain compatibility
    tfidf_df = pd.DataFrame(index=df.index)

# Note: The underlying issue, which caused many texts to become empty, is that
# the 'Customer Remarks' column was label-encoded to numerical values *before*
# text preprocessing was applied. This means the text processing functions were
# operating on numerical strings (e.g., '4253') rather than actual text.
# For truly meaningful text features, the text preprocessing steps should be
# performed on the original string content of 'Customer Remarks' *before*
# any numerical encoding of that column in your data wrangling section.
````

##### Which text vectorization technique have you used and why?

The technique used for vectorization is TF-IDF, or Term Frequency-Inverse Document Frequency.

TF-IDF is a technique for measuring the importance of a word in a document compared to the entire data set. It gives more weight to those words in a document that are more common in a specific text but rare in other texts. This technique is helpful in giving more importance to meaningful words while reducing the importance of common words.

The reason behind using this technique is that it can successfully measure the significance of the words in the customer message, which can help machine learning models better understand the data.

````python
# ML Model - 1 Implementation: Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize the model
lr_model = LogisticRegression(max_iter=1000, random_state=42)

# Fit the Algorithm on the resampled training data
lr_model.fit(X_train_res, y_train_res)

# Predict on the original test set
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
print("--- Logistic Regression Model Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Visualizing evaluation Metric Score chart for Logistic Regression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Confusion Matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues',
            xticklabels=lr_model.classes_, yticklabels=lr_model.classes_)
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report (Text output is usually sufficient, but can be visualized)
# For a chart, we can extract precision, recall, f1-score for each class
report_dict = classification_report(y_test, y_pred_lr, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()

# Drop accuracy, macro avg, weighted avg for cleaner visualization of per-class metrics
df_report = df_report.drop(labels=['accuracy', 'macro avg', 'weighted avg'], errors='ignore')

df_report[['precision', 'recall', 'f1-score']].plot(kind='bar', figsize=(10, 6))
plt.title('Precision, Recall, F1-Score per Class (Logistic Regression)')
plt.xlabel('CSAT Score Class')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Logistic Regression
param_grid_lr = {
    'C': [0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'], # 'liblinear' works well for small datasets and L1/L2, 'saga' for larger and L1/L2/Elastic-Net
    'penalty': ['l1', 'l2'] # L1 for 'liblinear' and 'saga', L2 for all
}

# Initialize GridSearchCV
grid_search_lr = GridSearchCV(
    estimator=LogisticRegression(max_iter=5000, random_state=42), # Increased max_iter for convergence
    param_grid=param_grid_lr,
    cv=5, # 5-fold cross-validation
    scoring='f1_macro', # Use f1_macro for imbalanced classification
    n_jobs=-1, # Use all available cores
    verbose=1
)

# Fit the Algorithm on the resampled training data
grid_search_lr.fit(X_train_res, y_train_res)

# Get the best estimator
best_lr_model = grid_search_lr.best_estimator_
print(f"Best Hyperparameters for Logistic Regression: {grid_search_lr.best_params_}")

# Predict on the original test set using the best model
y_pred_best_lr = best_lr_model.predict(X_test)

# Evaluate the best model
print("\n--- Optimized Logistic Regression Model Evaluation ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best_lr):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best_lr))
````
